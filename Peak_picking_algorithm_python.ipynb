```python
import matplotlib.pyplot as plt

def find_peaks_with_threshold(signal_vector, min_int_threshold, min_points):
    peaks = []
    peak_start = None
    peak_end = None
    peak_length = 0

    for index in range(len(signal_vector)):
        value = signal_vector[index]
        if value > min_int_threshold:
            if peak_start is None:
                peak_start = index
            peak_end = index
            peak_length += 1
        elif peak_start is not None:
            if peak_length >= min_points:
                peaks.append([peak_start, peak_end])
            peak_start = None
            peak_end = None
            peak_length = 0

    # Check if the last peak was not appended
    if peak_start is not None and peak_length >= min_points:
        peaks.append([peak_start, peak_end])

    return peaks

# Example usage
ints = [1.3, 1, 2, 4, 1.5, 1.8, 3, 5, 5.5, 4.3, 3, 1.5, 1.8, 1.4, 1.9, 1, 2, 3.2, 4.5, 5.3, 5.8, 5.1, 4, 3, 2, 1, 1.1]
min_int_threshold = 2.2
min_points = 2
peaks = find_peaks_with_threshold(ints, min_int_threshold, min_points)

# Plotting the signal and detected peaks
plt.plot(ints, label='Signal')
for peak in peaks:
    plt.axvspan(peak[0], peak[1], color='red', alpha=0.3)
plt.xlabel('Index')
plt.ylabel('Intensity')
plt.title('Detected Peaks')
plt.legend()
plt.show()
```
